```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np

cutoff = 1000
sample = 44102
```

```python
b    = signal.firwin(numtaps=5, cutoff=cutoff, fs=sample)
w, h = signal.freqz(b)

fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize=(15,7))
fig.suptitle('Digital filter frequency response', fontsize=15)

ax1.plot(w/np.pi*sample/2, 20 * np.log10(abs(h)), color='b', label='Amplitude')
ax1.set_ylabel('Amplitude [dB]', color='b')
ax1.set_xlabel('Frequency [Hz]')
ax1.grid(linestyle='dotted')
#ax1.set_xscale('log')
ax1.vlines(cutoff, min(20 * np.log10(abs(h))), max(20 * np.log10(abs(h))), linestyle='dashed')
#ax1.set_ylim(bottom=-17, top=1)


angles = np.unwrap(np.angle(h))
ax2.plot(w/np.pi*sample/2, angles, color='g', label='Phase')
ax2.set_xlabel('Frequency [Hz]')
ax2.set_ylabel('Angle (radians)', color='g')
ax2.grid(linestyle='dotted')
#ax2.vlines(cutoff, -200, 50, linestyle='dashed')
#ax2.set_ylim(bottom=-6.5, top=.5)
#ax2.axis('tight')
plt.show()
```

```python
b    = signal.firwin(numtaps=150, cutoff=cutoff, fs=sample, window=('kaiser', 8))
w, h = signal.freqz(b)

fig, ax1 = plt.subplots(figsize=(12,7))
ax1.set_title('Digital filter frequency response')
ax1.plot(w/np.pi*sample/2, 20 * np.log10(abs(h)), color='b', label='Amplitude')
ax1.set_ylabel('Amplitude [dB]', color='b')
ax1.set_xlabel('Frequency [rad/sample]')
ax1.set_xscale('log')
#ax1.vlines(25, -200, 50, linestyle='dashed')
ax1.set_ylim(bottom=-170, top=10)

ax2 = ax1.twinx()
angles = np.unwrap(np.angle(h))
ax2.plot(w/np.pi*sample/2, angles, color='g', label='Phase')
ax2.set_ylabel('Angle (radians)', color='g')
ax2.grid()
ax2.axis('tight')
plt.show()
```
